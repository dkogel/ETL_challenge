```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Extract
```

```python
#import CSVs
city_pollution = "../ETL_challenge/air_water_pollution.csv"
```

```python
city_info = pd.read_csv(city_pollution)
city_info.head()
```

```python
smart_city = "../ETL_challenge/Smart_City.csv"
```

```python
smart_info = pd.read_csv(smart_city)
smart_info.head()
```

```python
#Transform
```

```python
city_cols= ['City',' "Region"', ' "Country"', ' "AirQuality"',' "WaterPollution"']
```

```python
city_info_clean= city_info[city_cols].copy()
city_info_clean.head()
```

```python
city_info_clean= city_info_clean.rename(columns={'City':'city',' "Region"': 'region', ' "Country"': 'country',' "AirQuality"':'air_quality',' "WaterPollution"':'water_pollution'})
city_info_clean.head()
```

```python
smart_cols= ["City","Country","Smart_Mobility ","Smart_Environment","Smart_Living"]
smart_info_clean= smart_info[smart_cols].copy()
smart_info_clean.head()
```

```python
smart_info_clean= smart_info_clean.rename(columns={"City":"city","Country":"country","Smart_Mobility ":"smart_mobility",
                                                  "Smart_Environment":"smart_environment","Smart_Living":"smart_living"})
smart_info_clean.head()
```

```python
#Load
```

```python
# create database connection
connection_string = "postgres:bootcamp@localhost:5432/city_pollution_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# load dataframes into database
city_info_clean.to_sql(name='city_info', con=engine, if_exists='append', index=False)
```

```python
smart_info_clean.to_sql(name='smart_info', con=engine, if_exists='append', index=False)
```
