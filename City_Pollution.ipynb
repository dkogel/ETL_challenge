```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Extract
```

```python
#import CSVs
city_pollution = "../ETL_challenge/air_water_pollution.csv"
```

```python
city_info = pd.read_csv(city_pollution)
city_info.head()
```

```python
smart_city = "../ETL_challenge/Smart_City.csv"
```

```python
smart_info = pd.read_csv(smart_city)
smart_info.head()
```

```python
##Transform
```

```python
city_cols= ['City',' "AirQuality"',' "WaterPollution"']
```

```python
city_info_clean= city_info[city_cols].copy()
city_info_clean.head()
```

```python
city_info_clean= city_info_clean.rename(columns={'City':'city',' "AirQuality"':'air_quality',' "WaterPollution"':'water_pollution'})
city_info_clean.head()
```

```python
city_info_clean.set_index("city", inplace=True)
city_info_clean.head()
```

```python
smart_cols= ["City","Country","Smart_Mobility ","Smart_Environment","Smart_Living"]
smart_info_clean= smart_info[smart_cols].copy()
smart_info_clean.head()
```

```python
smart_info_clean= smart_info_clean.rename(columns={"City":"city","Country":"country","Smart_Mobility ":"smart_mobility",
                                                  "Smart_Environment":"smart_environment","Smart_Living":"smart_living"})
smart_info_clean.head()
```

```python
smart_info_clean.set_index("city", inplace=True)
smart_info_clean.head()
```
